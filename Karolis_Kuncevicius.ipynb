```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ei = 'Economy_Indicators.csv'
ei22 = pd.read_csv(ei)
ei22 = ei22.iloc[1:11]
```

```python
# Kokioje i TOP10 EU ali pagal BVP 2022m. buvo didiausias BVP?

fig, ax = plt.subplots()

width = 0.5
z = ax.bar(['Vokietija', 'Britanija', 'Pranczija', 'Italija', 'Rusija', 'Ispanija', 'Nyderlandai', 'Turkija', 'veicarija', 'Lenkija'], ei22['GDP'], width)
ax.set_ylabel('BVP Mlrd.')
ax.set_xlabel('alis')
ax.set_title('2022m. Europos ali ekonominiai rodikliai')
ax.bar_label(z, fmt='%.f')
plt.xticks(rotation = 65)
fig.tight_layout()

plt.show()
```

```python
# Kokioje i TOP10 EU ali pagal BVP 2022m. buvo didiausia infliacija?

fig, ax = plt.subplots()
w = 0.5
p = ax.bar(['Vokietija', 'Britanija', 'Pranczija', 'Italija', 'Rusija', 'Ispanija', 'Nyderlandai', 'Turkija', 'veicarija', 'Lenkija'], ei22['Inflation Rate'], width=w)
ax.set_ylabel('Infliacija %')
ax.set_xlabel('alis')
ax.set_title('2022m. Europos ali ekonominiai rodikliai')
ax.bar_label(p, fmt='%.2f')
plt.xticks(rotation = 65)
fig.tight_layout()

plt.show()
```

```python
# Kokioje i TOP10 EU ali pagal BVP 2022m. buvo didiausia skola/BVP?

fig, ax = plt.subplots()
w = 0.5
p = ax.bar(['Vokietija', 'Britanija', 'Pranczija', 'Italija', 'Rusija', 'Ispanija', 'Nyderlandai', 'Turkija', 'veicarija', 'Lenkija'], ei22['Debt/GDP'], width=w)
ax.set_ylabel('Skola/BVP')
ax.set_xlabel('alis')
ax.set_title('2022m. Europos ali ekonominiai rodikliai')
ax.bar_label(p, fmt='%.2f')
plt.xticks(rotation = 65)
fig.tight_layout()

plt.show()
```
